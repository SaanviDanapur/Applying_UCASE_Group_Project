## Exercise 1: Array Creation (5 Methods)

NumPy provides many ways to create arrays. Practice each method below.

```python
# Exercise 1: Create arrays using 5 different methods
import numpy as np
# Method 1: np.array() - create from a Python list
# TODO: Create an array from the list [10, 20, 30, 40, 50]
arr1 = np.array([10, 20, 30, 40, 50])
print("From list:", arr1)

# Method 2: np.zeros() - create array of zeros
# TODO: Create an array of 5 zeros
arr2 = np.zeros(5)
print("Zeros:", arr2)

# Method 3: np.ones() - create array of ones
# TODO: Create an array of 4 ones
arr3 = np.ones(4)
print("Ones:", arr3)

# Method 4: np.arange() - create a range of values
# TODO: Create an array from 0 to 9 (like range(10))
arr4 = np.arange(1, 11)
print("Range:", arr4)

# Method 5: np.linspace() - create evenly spaced values
# TODO: Create 5 evenly spaced values between 0 and 1
arr5 = np.linspace(0, 1, 5)
print("Linspace:", arr5)

# Hint: use np.array([...]), np.zeros(n), np.ones(n), np.arange(n), np.linspace(start, stop, num)
```

## Exercise 2: Element-Wise Math Operations

Unlike Python lists, NumPy arrays perform math on every element at once.

```python
import numpy as np
# Exercise 2: Math operations on arrays
prices = np.array([9.99, 14.99, 24.99, 4.99, 19.99])

# Task 1: Apply a 10% discount (multiply by 0.9)
# TODO: your code here
discounted = prices * 0.9
print("Discounted:", discounted)

# Task 2: Add $5 shipping to each price
# TODO: your code here
with_shipping = prices + 5
print("With shipping:", with_shipping)

# Task 3: Apply 8.25% sales tax (multiply by 1.0825)
# TODO: your code here
with_tax = prices * 1.825
print("With tax:", with_tax)

# Task 4: Add two arrays together
arr_a = np.array([1, 2, 3, 4, 5])
arr_b = np.array([10, 20, 30, 40, 50])
# TODO: your code here
combined = arr_a + arr_b
print("Combined:", combined)

# Hint: use standard arithmetic operators (+, *, /) directly on the array — NumPy applies them element-wise
```

## Exercise 3: 2D Arrays (Matrices)

2D arrays have rows and columns, like a spreadsheet or table.

```python
import numpy as np

# Exercise 3: Working with 2D arrays

# Create a 2D array (3 students, 4 test scores each)
grades = np.array([
    [85, 90, 78, 92],   # Student 1
    [88, 76, 95, 89],   # Student 2
    [92, 95, 88, 91]    # Student 3
])

# Task 1: Print the shape of the array
# TODO: your code here
print("Shape:", grades.shape)

# Task 2: Access Student 2's scores (row index 1)
# TODO: your code here
student_2 = grades[1]
print("Student 2 scores:", student_2)

# Task 3: Access all students' first test score (column index 0)
# TODO: your code here
test_1 = grades[:,0]
print("Test 1 scores:", test_1)

# Task 4: Access Student 3's third test score (row 2, col 2)
# TODO: your code here
single_score = grades[2][2]
print("Student 3, Test 3:", single_score)

# Hint: use grades[row_index] for a full row, grades[:, col_index] for a column, grades[r, c] for one value
```

## Exercise 4: Array Slicing

Slicing lets you select a portion of an array using `[start:stop:step]`.

```python
import numpy as np
# Exercise 4: Slicing practice
data = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50])
print("Full array:", data)

# Task 1: Get elements at index 1 through 4 (10, 15, 20, 25)
# TODO: your code here
slice1 = data[1:5]
print("Index 1-4:", slice1)

# Task 2: Get the first 3 elements (5, 10, 15)
# TODO: your code here
slice2 = data[0:3]
print("First 3:", slice2)

# Task 3: Get the last 3 elements (40, 45, 50)
# TODO: your code here
slice3 = data[-3:]
print("Last 3:", slice3)

# Task 4: Get every other element (5, 15, 25, 35, 45)
# TODO: your code here
slice4 = data[::2]
print("Every other:", slice4)

# Task 5: Reverse the array
# TODO: your code here
slice5 = data[::-1]
print("Reversed:", slice5)

# Hint: use Python slice notation [start:stop:step]
```

## Exercise 5: Boolean Indexing

Boolean indexing lets you filter arrays based on conditions — like a WHERE clause in SQL!

```python
import numpy as np
# Exercise 5: Boolean indexing
temperatures = np.array([68, 72, 85, 90, 55, 78, 95, 62, 88, 73])
print("Temperatures:", temperatures)

# Task 1: Find all temperatures above 80
# TODO: your code here
hot_days = temperatures > 80
print("Hot days (>80):", hot_days)

# Task 2: Find all temperatures between 70 and 85 (inclusive)
# Hint: use & for AND, and wrap each condition in parentheses
# TODO: your code here
comfortable = (temperatures >= 70) & (temperatures <= 85)
print("Comfortable (70-85):", comfortable)

# Task 3: Count how many days were below 65
# Hint: boolean array .sum() counts True values
# TODO: your code here
cold_count = np.sum(temperatures < 65)
print("Cold days count:", cold_count)

```

## Exercise 6: Aggregate Functions

NumPy makes it easy to compute statistics on your data.

```python
# Exercise 6: Aggregate statistics
exam_scores = np.array([78, 85, 92, 67, 90, 88, 73, 95, 82, 79])
print("Exam scores:", exam_scores)

# Task 1: Calculate the mean (average)
# TODO: your code here
mean_score = np.mean(exam_scores)
print(f"Mean: {mean_score}")

# Task 2: Calculate the sum
# TODO: your code here
total = np.sum(exam_scores)
print(f"Sum: {total}")

# Task 3: Find the maximum score
# TODO: your code here
highest = np.max(exam_scores)
print(f"Max: {highest}")

# Task 4: Find the minimum score
# TODO: your code here
lowest = np.min(exam_scores)
print(f"Min: {lowest}")

# Task 5: Calculate the standard deviation
# TODO: your code here
spread = np.std(exam_scores)
print(f"Std Dev: {spread:.2f}")

# Task 6: Print a summary using f-strings
# TODO: Print "The class averaged {mean}, with scores ranging from {min} to {max}"


# Hint: use sum() and len() for a manual average, or np.mean() / .mean()
```
