## Exercise 2: `.describe()` — Summary Statistics

`.describe()` gives you count, mean, std, min, 25%, 50%, 75%, max for every numeric column.

- **count**: number of non-null values

- **min/max**: smallest/largest values
- **25%/50%/75%**: quartiles (50% = median)

```python
import pandas as pd
import matplotlib.pyplot as plt
# Exercise 2: Use .describe() and interpret the output
df_emp = pd.read_csv('https://raw.githubusercontent.com/zsmithc2c/dam-course/main/datasets/phase_1/employee_data.csv')
# Task 1: Run describe() on the employee DataFrame
# TODO: your code here
df_emp.describe()

# Task 2: Answer these questions in comments:
# What is the average salary?
# What is the salary range (max - min)?
# What does the 50% (median) tell you compared to the mean?

# TODO: Write your answers as comments
# Average salary: $72720
# Salary range: $74000
# Median vs mean tells me: The middle value of the salary value rather than the overall average

# Hint: call .describe() on the DataFrame
```

## Exercise 4: `.value_counts()`

`.value_counts()` counts how often each unique value appears in a column. Perfect for categorical data.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Exercise 4: value_counts on categorical columns
df_emp = pd.read_csv('https://raw.githubusercontent.com/zsmithc2c/dam-course/main/datasets/phase_1/employee_data.csv')
# Task 1: Count how many employees are in each department
# TODO: your code here
dept_counts = df_emp.value_counts()
print("Department counts:")
print(dept_counts)
print()

# Task 2: If there's a 'gender' or 'role' column, count those values too
# TODO: your code here (pick any categorical column)
df_emp['gender'].value_counts()
# Hint: use .value_counts() on a Series column
```

## Exercise 5: `.sort_values()`

Sort your data by any column, ascending or descending.

```python
import pandas as pd
import matplotlib.pyplot as plt
# Exercise 5: Sorting
df_emp = pd.read_csv('https://raw.githubusercontent.com/zsmithc2c/dam-course/main/datasets/phase_1/employee_data.csv')
# Task 1: Sort employees by salary (highest first)
# TODO: your code here
sorted_df = df_emp.sort_values('salary', ascending=False)
print("Top earners:")
print(sorted_df.head())
print()

# Task 2: Sort employees by name (alphabetical)
# TODO: your code here
alpha_df = df_emp.sort_values('name')
print("Alphabetical:")
print(alpha_df.head())

# Hint: review the instructions above and apply the relevant function
```
